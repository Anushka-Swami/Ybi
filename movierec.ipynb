# **Get Feature test Conversion to Tokens**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
x = tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

# **Get similarity score using cosine similarity**

Compute cosine similarity between samples in X and Y.

Cosine similarity, or the cosine kernel, computes similarity as the normalized dot product of X and Y:

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_score=cosine_similarity(x)
```

```python
Similarity_score
```

```python
Similarity_score.shape
```

Get movie name as the input from the user and validate for closest spelling

```python
Favourite_Movie_Name= input('Enter your favourite movie name:')
```

```python
All_Movies_title_List=data['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_Recommendation = difflib.get_close_matches(Favourite_Movie_Name,All_Movies_title_List)
print(Movie_Recommendation)
```

```python
Close_Match = Movie_Recommendation[0]
print(Close_Match)
```

```python
Index_of_close_match = data[data.Movie_Title==Close_Match]['Movie_ID'].values[0]
print(Index_of_close_match)
```

```python
#getting list of similar movies
Recommendation_Score= list(enumerate(Similarity_score[Index_of_close_match]))
print(Recommendation_Score)
```

```python
len(Recommendation_Score)
```

Get all movies sort based on their similaity score

```python
Sorted_Similar_Movies= sorted(Recommendation_Score,key=lambda x:x[1], reverse= True)
print(Sorted_Similar_Movies)
```

```python
print('Top #30 Movies Sugggested or you: \n')
i=1
for movie in Sorted_Similar_Movies:
  index=movie[0]
  title_from_index=data[data.index==index]['Movie_Title'].values[0]
  if(i<31):
    print(i,',',title_from_index)
    i+=1
```
